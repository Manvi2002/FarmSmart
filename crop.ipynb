```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#load dataset
df=pd.read_csv("Crop_recommendation.csv")
df.head()
```

```python
df.label.unique()
```

```python
#preprocessing of data
df.describe()
```

```python
df.info()
```

features: ph,N,P,K,rainfall,temperature & humidity
target->**label** **bold text**

```python
#encoding
le=LabelEncoder()
df['label']=le.fit_transform(df['label'])
df["label"]
```

```python
#label mapping
for index,label in enumerate(le.classes_):
  print(f"{label}->{index}")
```

```python
#imbalance
df.label.value_counts()
```

```python
df.head()
```

```python
scaler=StandardScaler()
scaler_values=scaler.fit_transform(df[['temperature','humidity','ph','rainfall']])
```

print(scaler_values)

```python
#array->df
scaled_df=pd.DataFrame(scaler_values,columns=['temperature','humidity','ph','rainfall'])
#reseting index
scaled_df.index=df.index
#replace scaled df to original df
df[['temperature','humidity','ph','rainfall']]=scaled_df
```

```python
df.head()
```

```python
#check maps
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('Correlation HeatMap')
plt.show()
```

creating **Model**

```python
#splitting dataset infeatures and target var
x=df[['N','P','K','temperature','humidity','ph','rainfall']]
y=df['label']
```

```python
#split train test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
#train model
model=RandomForestClassifier(random_state=45)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(f"Classification report->\n{classification_report(y_test,y_pred)}")
print(f"Accuracy->{accuracy_score(y_test,y_pred)}")
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.title('Confusion Matrix Heatmap')
plt.show()
```
